The primary dataset aids in visualizing the popularity and energy of Taylor Swift's songs categorized by album. I found the csv file at https://www.kaggle.com/thespacefreak/taylor-swift-spotify-data/version/1, which allows anyone to download by logging in with your google account. This dataset is named "spotify_taylorswift.csv" and highlights the overall popularity, energy, danceability, loudness, etc. of each of Taylor Swift's songs from each of her produced albums. I found this to be very interesting, because this artist is so widely known by young women and her success has grown through each album. I named this dataset as "swift". I believe the source must have included spotify in their title of the file because that may have been where they collected data on 'popularity' and 'length' of each song.

```python
swift = pd.read_csv('https://raw.githubusercontent.com/mke4/mke4.github.io/main/spotify_taylorswift.csv')
swift
```

```python
albums = swift['album'].unique()
label = ipywidgets.Dropdown(options = albums, value = 'reputation', description='Album:')
initial_album = label.value
initial_song = swift['name'][swift['album']== label.value]
initial_pop = swift['popularity'][swift['album']== label.value]
initial_energy = swift['energy'][swift['album']== label.value]
```

```python
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.LinearScale()

x_ax = bqplot.Axis(scale = x_sc, label='song', tick_rotate=45, label_offset='50')
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical', 
                   label='energy')


# now replot:
barp = bqplot.Bars(x = initial_song,
                          y = initial_energy,
                          scales = {'x': x_sc, 'y': y_sc})


def update_fig(*args):
    initial_album = label.value
    barp.x = swift['name'][swift['album']== label.value]
    barp.y = swift['energy'][swift['album']== label.value]

label.observe(update_fig, 'value')
fig = bqplot.Figure(marks = [barp], axes = [y_ax, x_ax])
fig.axes[1].tick_style = {'text-anchor': 'start', 'font-size':5}


myDashboard = ipywidgets.VBox([label, fig])
myDashboard
```

```python
swift_yt = pd.read_csv('https://raw.githubusercontent.com/mke4/mke4.github.io/main/taylor_channel.csv')
swift_yt
```

source for how to extract only the year https://www.codegrepper.com/code-examples/python/HOW+TO+CONVERT+THE+DATE+COLUMN+ONLY+YEAR+PYTHON

```python
swift_yt['publication_date']= pd.to_datetime(swift_yt['publication_date'])
swift_yt['year']= pd.DatetimeIndex(swift_yt['publication_date']).year
swift_yt['year']
```

```python
swift['release_date']= pd.to_datetime(swift['release_date'])
swift['year']= pd.DatetimeIndex(swift['release_date']).year
swift['year']
```

```python
x_scg = bqplot.LinearScale()
y_scg = bqplot.LinearScale()

x_axg = bqplot.Axis(scale = x_scg, label = 'year')
y_axg = bqplot.Axis(scale = y_scg, label = 'song likes', label_offset='-40', orientation = 'vertical', tick_format = '.2s')

bar_yt = bqplot.Bars(x=swift_yt['year'], y=swift_yt['likeCount'], scales={'x':x_scg, 'y':y_scg})

fig_yt = bqplot.Figure(marks = [bar_yt], axes = [y_axg, x_axg])
fig_yt
```

```python
x_scj = bqplot.LinearScale()
y_scj = bqplot.LinearScale()

x_axj = bqplot.Axis(scale = x_scj, label = 'year')
y_axj = bqplot.Axis(scale = y_scj, label = 'likes/popularity', label_offset='-40', orientation = 'vertical', tick_format = '.2s')
                    
                    
bar_yts = bqplot.Bars(x=swift_yt['year'], y=[swift_yt['likeCount'], swift_yt['dislikeCount']], scales={'x':x_scj, 'y': y_scj}, type="grouped", 
                      labels=["Likes", "Dislikes"], display_legend=True, 
                      colors=["green", "red"])
            
fig_yts = bqplot.Figure(marks=[bar_yts], axes=[x_axj, y_axj], legend_style={"fill":"white"}, title="Spotify versus YouTube popularity")
fig_yts
```
